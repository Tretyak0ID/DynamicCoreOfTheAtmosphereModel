```python
basis = 2
num   = 9

A = SA.ApproxAnalyzer('Euler', 'LeftPeriodic')
l2Error = A.GaussApprox(basis,num)

p = np.zeros(num-1)
ph = np.zeros(num-1)

for i in range(num-1):
    ph[i] = i
    p[i]  = log2(l2Error[i]/l2Error[i+1])
    
plt.plot(ph,p)
plt.plot(ph, np.ones(ph.size))
```
