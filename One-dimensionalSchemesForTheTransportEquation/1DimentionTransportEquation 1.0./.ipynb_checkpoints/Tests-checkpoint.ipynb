```python
#-ПОДКЛЮЧЕНИЕ-МОДУЛЕЙ-И-БИБЛИОТЕК-
%matplotlib notebook

import numpy              as np
import scipy
import matplotlib.pyplot  as plt
import glob
import moviepy.editor     as mpy
import nbimporter
import CyclicScheme1TransportEquation as schemes

from scipy                import sparse
from scipy.sparse         import linalg
from numpy                import pi, sin, cos, ma
from pylab                import *
from mpl_toolkits.mplot3d import Axes3D
from matplotlib           import cm
from matplotlib.ticker    import LinearLocator, FormatStrFormatter
from matplotlib           import animation
from numpy.linalg         import norm
```

```python
x_max   = 2*pi
t_max   = 2*pi
count_t = 200
count_x = 100
h = x_max/count_x

SC = sin(np.arange(0, (count_x+1), dtype=double)*h)
[x,t,PHI] = schemes.CyclicScheme1TransportEquation('RK4+SBP-4,2+SAT', x_max, t_max, count_x, count_t, 1, SC)
```

```python
#-ВИЗУАЛИЗАЦИЯ-
fig, ax = plt.subplots()
plt.style.use('dark_background')
surf = ax.plot(x,PHI[0])

def animate(n):
    plt.style.use('dark_background')
    ax.clear()
    ax.set_ylim(-2,2)
    surf = ax.plot(x,PHI[n])
    return surf

anim = animation.FuncAnimation(fig, animate, frames=len(t), interval=1, repeat = False)
```
